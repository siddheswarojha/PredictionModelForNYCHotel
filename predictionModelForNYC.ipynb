```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import joblib
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df
```

```python
df1 = df.drop(['host_id','host_name','latitude','longitude','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count'],axis='columns')
df1
```

```python
df1.minimum_nights.isnull().sum()
```

```python
df1.availability_365.isnull().sum()
```

```python
df1.neighbourhood_group.unique()
```

```python
len(df1.name.unique())
```

```python
df1['price_per_night']=df1['price']/df1['minimum_nights']
df1
```

```python
df1.describe()
```

```python
min_threshold= df1['price_per_night'].quantile(0.5)

min_threshold
```

```python
max_threshold = df1['price_per_night'].quantile(0.999)

max_threshold
```

```python
df2 = df1[(df1.price_per_night>min_threshold)&(df1.price_per_night<max_threshold)]
df2
```

```python
newDF = pd.get_dummies(df2.neighbourhood_group)
newDF
```

```python
bestDF = pd.concat([df2,newDF],axis='columns')
bestDF
```

```python
len(bestDF.name.unique())
```

```python
len(bestDF.room_type.unique())
```

```python
new_best_DF = pd.get_dummies(bestDF.room_type)
new_best_DF
```

```python
finalDF = pd.concat([bestDF,new_best_DF],axis='columns')
finalDF
```

```python
X= finalDF.drop(['name','neighbourhood_group','neighbourhood','room_type','price'],axis='columns')
X
```

```python
y=finalDF.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.01)
```

```python
predictionModel = linear_model.LinearRegression()
predictionModel.fit(X_train,y_train)
```

```python
predictionModel.score(X_test,y_test)
```

```python
joblib.dump(predictionModel,"predictionModelForNYCHotel")
```
